```python
# Part 3 import and prep data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

focuscrimes = ['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT']

data = pd.read_csv("/Users/sorenbendtsen/Documents/DTU Business Analytics/1st Semester/Social Data Analysis and Visualization/F1Exercises/Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")
# Convert Date to a datetime format
data["Date"] = pd.to_datetime(data["Date"])
# Make a subset without 2018, as we only look on complete years.
data = data[data["Date"].dt.year != 2018]
# Create year and month variable.
data["Year"] = data["Date"].dt.year
data["Month"] = data["Date"].dt.month
data["HourInt"] = pd.to_datetime(data["Time"], format = '%H:%M').dt.hour
data["Minute"] = pd.to_datetime(data["Time"], format = '%H:%M').dt.minute
data["Weekday"] = pd.to_datetime(data.Date).dt.weekday
data["Week_hour"] = data.Weekday*24 + data.HourInt
data = data[data["Category"].isin(focuscrimes)]
data.head()
```

```python
# Part 3 Exercise 1 - jitter plot
import seaborn as sns
# Filter data on one category and six months in a specific hour
data_jitter=data[data['Category'] == "ROBBERY"]
data_jitter=data_jitter.loc[(data_jitter['Date'] >= '2010-01-01') & (data_jitter['Date'] < '2010-07-01')]
data_jitter=data_jitter.loc[(data_jitter['Time'] >= '13:00') & (data_jitter['Time'] < '14:00')]
data_jitter = data_jitter.sort_values(by='Time')

axes = sns.stripplot(data=data_jitter, x='Time', marker='$\circ$', s=10)
axes.tick_params(axis='x', rotation=90)
#
```

```python
# Exercise 1 - histograms

# Filter data for 2 categories and 1 year.
data_hist = data[(data['Category'] == 'PROSTITUTION') | (data['Category'] == 'TRESPASS')]
data_hist = data_hist.loc[(data_hist['Date'] >= '2017-01-01') & (data_hist['Date'] < '2018-01-01')]
data_hist.shape

# Plot histogram of latitude (Y) with 50 bins.
# Creating plot
fig = plt.figure(figsize =(18, 8))

n_bins = 50 
plt.hist(data_hist['Y'], bins = n_bins, edgecolor='black', facecolor ='None', lw=1)
plt.title("Numpy Histogram of Latitudes", fontsize=20, weight='bold')
plt.xticks([37.71, 37.72, 37.73, 37.74, 37.75, 37.76, 37.77, 37.78, 37.79, 37.80, 37.81])
plt.xlabel('Latitude', fontsize=20)
plt.ylabel('Number of crimes', fontsize=20)
 
# show plot
plt.show()
```

```python
# Part 4 exercise 1 - use plotly to create a choropleth mapbox based on san francisco district json data and randomdata
import plotly.express as px

from urllib.request import urlopen
import json
with urlopen('https://raw.githubusercontent.com/suneman/socialdata2022/main/files/sfpd.geojson') as response:
    counties = json.load(response)

counties["features"][0]

fig = px.choropleth_mapbox(randomdata, geojson=counties, locations='District', color='Crime',
                           color_continuous_scale="Viridis",
                           range_color=(0, 1),
                           mapbox_style="carto-positron",
                           zoom=10.5, center = {"lat": 37.773972, "lon": -122.431297},
                           opacity=0.5,
                           labels={'crime':'number of crimes'}
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
# Part 4 Exercise 2 - crime data visualization
# Category = VEHICLE THEFT and DayOfWeek = Sunday.
# Change color scale to OrRd and range to min and max of crime variable.
data_viz_car = data[(data['Category'] == 'VEHICLE THEFT') & (data['DayOfWeek'] == 'Sunday')]
data_viz_car = data_viz_car.groupby('PdDistrict')['PdId'].count().reset_index(name='Vehicle_Theft')

fig = px.choropleth_mapbox(data_viz_car, geojson=counties, locations='PdDistrict', color='Vehicle_Theft',
                           color_continuous_scale="OrRd",
                           range_color=(data_viz_car.Vehicle_Theft.min(), data_viz_car.Vehicle_Theft.max()),
                           mapbox_style="carto-positron",
                           zoom=10.5, center = {"lat": 37.773972, "lon": -122.431297},
                           opacity=0.5,
                           labels={'Vehicle Theft':'number of crimes'}
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()

# From the figure below, it is clearly visualized that the worse area to park on a Sunday is in Ingleside, 
# where the best area to park is in Tenderloin.


```

```python
# Try to make an interactive dashboard using dash for a dropdown menu of weekdays.
data_viz_car_week = data[data['Category'] == 'VEHICLE THEFT']
data_viz_car_week = data_viz_car_week.groupby(['DayOfWeek', 'PdDistrict'])['PdId'].count().reset_index(name='Vehicle_Theft')

from dash import Dash, dcc, html, Input, Output

app = Dash(__name__)

app.layout = html.Div([
    html.H4('Concentration of vehicle theft in San Francisco'),
    html.P("Select the day of week:"),
    dcc.RadioItems(
        id='dropdown', 
        options=data_viz_car_week['DayOfWeek'],
        value="Monday",
        inline=True
    ),
    dcc.Graph(id="graph"),
])

@app.callback(
    Output("graph", "figure"), 
    Input("dropdown", "value"))
def display_choropleth(dayofweek):
    filtered_data = data_viz_car_week[data_viz_car_week['DayOfWeek'] == dayofweek]

    fig = px.choropleth_mapbox(filtered_data, geojson=counties, locations='PdDistrict', color='Vehicle_Theft',
                           color_continuous_scale="OrRd",
                           range_color=(filtered_data.Vehicle_Theft.min(), filtered_data.Vehicle_Theft.max()),
                           mapbox_style="carto-positron",
                           zoom=10.5, center = {"lat": 37.773972, "lon": -122.431297},
                           opacity=0.5,
                           labels={'Vehicle Theft':'number of crimes'}
                          )
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()

app.run_server(debug=True, use_reloader=False)
```
