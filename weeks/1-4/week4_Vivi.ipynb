```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
data = pd.read_csv('crime.csv')
```

```python
columns = ['Category', 'Date', 'Time', 'DayOfWeek', 'Address', 'X', 'Y']
crimeData = pd.DataFrame(data, columns=columns)
crimeData.rename(columns={'X': 'longitude', 'Y': 'latitude'}, inplace=True)
crimeData['Date'] = pd.to_datetime(crimeData['Date'])
crimeData['Time'] = pd.to_datetime(crimeData['Time'])
crimeData['latitude'] = crimeData['latitude'].astype(float)
crimeData['longitude'] = crimeData['longitude'].astype(float)
crimeData
```

```python
sf_map = folium.Map([37.7749, -122.4192], tiles = "Stamen Toner", zoom_start=13)
# 37.77919, -122.41914
folium.Marker([37.77919, -122.41914], popup='City Hall').add_to(sf_map)
sf_map
```

```python
drugData = crimeData[crimeData['Category'] == 'DRUG/NARCOTIC']
drugData.set_index('Date')
drugData = drugData.query('20160601 <= Date <= 20160731')
drugData
```

```python
drugMap = folium.Map([37.7749, -122.4192], tiles = "Stamen Toner", zoom_start=13)

def plotDot(point):
    folium.CircleMarker(location=[point.latitude, point.longitude],
                        radius=2).add_to(drugMap)

drugData.apply(plotDot, axis=1)
#drugMap.fit_bounds(drugMap.get_bounds())
drugMap
```

```python
sexCrime = crimeData.query("Category == 'SEX OFFENSES, NON FORCIBLE'")
sexCrime
```

```python
heat_data = [[row['latitude'],row['longitude']] for index, row in sexCrime.iterrows()]
heatMap = folium.Map([37.7749, -122.4192], tiles = "Stamen Toner", zoom_start=13)
HeatMap(heat_data, radius=50, blur=30).add_to(heatMap)
heatMap
```

```python
drunkCrime = crimeData[crimeData['Category'] == 'DRIVING UNDER THE INFLUENCE']
drunkCrime = drunkCrime[drunkCrime['Date'].dt.year == 2016]
drunkCrime = drunkCrime[drunkCrime['Date'].dt.month == 12]
drunkCrime['Day'] = drunkCrime['Date'].dt.day
drunkCrime
```

```python
heat_data = [[[row['latitude'],row['longitude']] for index, row in drunkCrime[drunkCrime['Day'] == i].iterrows()] for i in range(0,32)]
drunkMap = folium.Map([37.7749, -122.4192], tiles = "Stamen Toner", zoom_start=13)
hm = folium.plugins.HeatMapWithTime(heat_data,auto_play=False,max_opacity=0.8)
hm.add_to(drunkMap)
drunkMap
```
