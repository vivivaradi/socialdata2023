```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
data = pd.read_csv('crime.csv')
data['Date'] = pd.to_datetime(data['Date'])
data
```

```python
theft = data.query('Category == "LARCENY/THEFT" & Resolution == "ARREST, CITED" & Date.dt.year == 2013 & Date.dt.month >= 3 & Date.dt.month < 9')
theft
```

```python
hour = theft.query('Time >= "13:00" & Time < "14:00"')
hour = hour.sort_values(by='Time')
hour
```

```python
axes = sns.stripplot(data=hour, x='Time', marker='$\circ$', s=10)
axes.tick_params(axis='x', rotation=90)
```
